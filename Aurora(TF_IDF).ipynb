```python
from docx import Document
def read_events_from_docx():
    doc = Document("/content/12123.docx")
    events_dict = {}
    current_event = None

    for para in doc.paragraphs:
        text = para.text.strip()
        if not text:
            continue


        if text[0].isdigit() and "." in text:
            current_event = text.split(". ", 1)[1]
            events_dict[current_event] = []
        elif current_event:
            events_dict[current_event].append(text)

    return events_dict


events_dict = read_events_from_docx()
for event, items in events_dict.items():
    print(f"{event}: {items}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Векторизація назв товарів

```python
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(stocks_df["Назва"])
```

```python
def find_similar_products(query, top_n=5):
    query_vec = vectorizer.transform([query])
    similarity_scores = cosine_similarity(query_vec, tfidf_matrix).flatten()

    top_indices = similarity_scores.argsort()[-top_n:][::-1]
    return stocks_df.iloc[top_indices][["Назва", "Підкатегорія", "Залишок_Товарів_КінецьДня"]]

```

```python
def recommend_products_for_event(event_name, top_n=5):
    if event_name not in events_dict:
        return f"Подію '{event_name}' не знайдено в базі."

    recommended_items = []
    for keyword in events_dict[event_name]:
        result = find_similar_products(keyword, top_n=3)
        recommended_items.append(result)
    combined_results = pd.concat(recommended_items)
    combined_results = combined_results.drop_duplicates().head(top_n)

    return combined_results
```

```python
event = "Вечір читання"
recommendations = recommend_products_for_event(event, top_n=5)
```

```python
print(f"Рекомендації для {event}:")
print(recommendations.to_string(index=False))
```
